### Tokenizing the Data

We use TensorFlow's `Tokenizer` class to convert the raw text data into sequences of integers, where each integer corresponds to a unique word in the dataset.


```python
#Tokenize the data using the tensorflow tokenizer
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer

tokenizer = Tokenizer()
tokenizer.fit_on_texts([data])

tokenizer.word_index
```

### Creating Input Sequences

We begin by splitting the text data into individual sentences. Each sentence is then tokenized into a sequence of integers. To prepare the data for the model, we generate input sequences of increasing length by progressively appending tokens from each sentence. This allows the model to learn from varying sequence lengths and enables sequential prediction during training. The resulting `input_sequences` list will contain all these generated tokenized sequences.


```python
input_sequences = []
for sentence in data.split('\n'):
  tokenized_sentence = tokenizer.texts_to_sequences([sentence])[0]

  for i in range(1,len(tokenized_sentence)):
    input_sequences.append(tokenized_sentence[:i+1])


print(input_sequences)
```

```python
#Max length of my input sequences
max_len = max([len(x) for x in input_sequences])
```

```python
print(max_len)
```

```python
len(input_sequences)
```

```python
#Pad the sequences to make all the sequences into same size

from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input_sequences = pad_sequences(input_sequences,maxlen = max_len,padding='pre')
```

```python
#Making My input and output data

x = padded_input_sequences[:,:-1]
y = padded_input_sequences[:,-1]
```

```python
#Converting my output data into a categorical data
from tensorflow.keras.utils import to_categorical
y = to_categorical(y,num_classes = 4295)
```

### Building the LSTM Model

We create a Sequential model with the following layers:

1. **Embedding Layer**: Represents words as dense vectors of size 100, with a vocabulary size of 4295.
2. **LSTM Layer**: 150 units to capture sequential patterns in the text.
3. **Dense Layer**: 4295 units with `softmax` activation for predicting the next word.

The model is compiled with `categorical_crossentropy` loss and the `adam` optimizer, and accuracy is used as the evaluation metric.


```python
# building the model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,LSTM,Embedding

model = Sequential()

model.add(Embedding(4295,100))
model.add(LSTM(150))
model.add(Dense(4295,activation = 'softmax'))

model.compile(loss= 'categorical_crossentropy',optimizer = 'adam',metrics = ['accuracy'])

model.summary()
```

```python
#Train my model
model.fit(x,y,epochs = 100)
```

### Prediction System

The following steps are taken to generate the next words in the sequence:

1. **Input Sequence**: The starting text ("The truth is") is tokenized and padded.
2. **Word Prediction**: The model predicts the next word based on the given input sequence.
3. **Update Text**: The predicted word is added to the input text, and the process is repeated for 10 iterations.
4. **Sleep**: A 2-second delay is added after each word prediction to simulate a real-time prediction.

This process results in the auto-generation of a text sequence based on the model's predictions.


```python
# Prediction System
import numpy as np
import time

text = 'The truth is'

for i in range(10):

  tokenize_text = tokenizer.texts_to_sequences([text])[0]

  padded_input_text = pad_sequences([tokenize_text],maxlen = max_len-1,padding = 'pre')

  pos = np.argmax(model.predict(padded_input_text))

  for word,index in tokenizer.word_index.items():
    if index == pos:
      text = text + " " + word
      print(text)
      time.sleep(2)
```
